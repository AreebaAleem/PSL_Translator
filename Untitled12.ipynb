```python
import cv2
import os

def extract_frames_debug(video_path, save_path, num_frames=30):
    """
    Extracts frames from a video and saves them to the specified folder.
    Args:
    - video_path: Path to the video file.
    - save_path: Path where extracted frames will be saved.
    - num_frames: Number of frames to extract from the video.
    """
    cap = cv2.VideoCapture(video_path)

    # Check if the video was opened successfully
    if not cap.isOpened():
        print(f"Error: Cannot open video {video_path}")
        return

    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    print(f"Total frames in {video_path}: {total_frames}")

    interval = total_frames // num_frames
    os.makedirs(save_path, exist_ok=True)

    frame_count = 0
    extracted_count = 0

    for i in range(num_frames):
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_count)
        ret, frame = cap.read()

        if ret:
            frame_path = f"{save_path}/frame_{i}.jpg"
            cv2.imwrite(frame_path, frame)
            print(f"Saved: {frame_path}")
            extracted_count += 1
        else:
            print(f"Error: Couldn't read frame {i} from {video_path}")

        frame_count += interval

    cap.release()
    print(f"Extracted {extracted_count} frames from {video_path}.")

# Example: Extract frames for videos in the 'stiff' gesture folder and save to Google Drive
gesture_folder = '/content/validate/validate/surprise/'  # Adjust this path to your local folder
output_folder = '/content/drive/MyDrive/extracted_frames/validate/surprise/'  # Google Drive destination

# Ensure the folder in Google Drive exists
os.makedirs(output_folder, exist_ok=True)

# Process all videos in the folder and save frames to Google Drive
for video_file in os.listdir(gesture_folder):
    video_path = os.path.join(gesture_folder, video_file)
    extract_frames_debug(video_path, f"{output_folder}/{os.path.splitext(video_file)[0]}")
```

```python
import numpy as np
import os
import cv2
import mediapipe as mp
import pickle
```

```python
# Initialize MediaPipe
mp_hands = mp.solutions.hands
mp_pose = mp.solutions.pose
hands = mp_hands.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.5)
pose = mp_pose.Pose(static_image_mode=True)
```

```python
def extract_keypoints(image_path):
    image = cv2.imread(image_path)
    if image is None:
        return np.zeros(75)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Extract hand landmarks
    hand_results = hands.process(image_rgb)
    hand_keypoints = np.zeros(63)
    if hand_results.multi_hand_landmarks:
        hand_keypoints = np.array([[lm.x, lm.y, lm.z] for lm in hand_results.multi_hand_landmarks[0].landmark]).flatten()

    # Extract pose landmarks
    pose_results = pose.process(image_rgb)
    pose_keypoints = np.zeros(12)
    if pose_results.pose_landmarks:
        pose_keypoints = np.array([[pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].z],
                                   [pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].z],
                                   [pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].z],
                                   [pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].z]]).flatten()

    return np.concatenate([hand_keypoints, pose_keypoints])
```

```python
import os
import numpy as np
import time

def prepare_data(folder_path, save_path='/content/drive/MyDrive/processed_data', save_interval=100):
    keypoints_data = []
    labels = []
    total_images = 0  # To keep track of total processed images

    os.makedirs(save_path, exist_ok=True)  # Ensure the save path exists

    for gesture_folder in os.listdir(folder_path):
        gesture_path = os.path.join(folder_path, gesture_folder)
        if not os.path.isdir(gesture_path):
            continue

        for subdir in os.listdir(gesture_path):
            subdir_path = os.path.join(gesture_path, subdir)
            if not os.path.isdir(subdir_path):
                continue

            image_count = 0
            for frame in os.listdir(subdir_path):
                frame_path = os.path.join(subdir_path, frame)
                if frame_path.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                    keypoints = extract_keypoints(frame_path)
                    if keypoints.size > 0:
                        keypoints_data.append(keypoints)
                        labels.append(gesture_folder)
                        image_count += 1

            print(f"Processed {image_count} images from {subdir_path}")
            total_images += image_count

            # Save progress periodically with unique filenames to avoid overwriting
            if total_images % save_interval == 0:
                np.save(os.path.join(save_path, f'keypoints_data_{total_images}.npy'), np.array(keypoints_data))
                np.save(os.path.join(save_path, f'labels_{total_images}.npy'), np.array(labels))
                print(f"Progress saved at {total_images} images")

            time.sleep(1)  # Optional sleep to keep session alive

    # Final save using main filenames
    np.save(os.path.join(save_path, 'keypoints_data.npy'), np.array(keypoints_data))
    np.save(os.path.join(save_path, 'labels.npy'), np.array(labels))
    print(f"Final save complete. Feature shape: {len(keypoints_data)}, Label shape: {len(labels)}")

    # **Verification Check**: Load the saved arrays right after the final save
    keypoints_test = np.load(os.path.join(save_path, 'keypoints_data.npy'))
    labels_test = np.load(os.path.join(save_path, 'labels.npy'))
    print(f"Loaded Feature shape: {keypoints_test.shape}, Loaded Labels shape: {labels_test.shape}")

# Usage
train_folder = '/content/drive/MyDrive/extracted_frames/train'
prepare_data(train_folder)
```

```python
import matplotlib.pyplot as plt # import the pyplot submodule

def plot_keypoints(keypoints, label):
    num_keypoints = 25  # Adjust based on your interpretation
    x_coords = keypoints[:num_keypoints]  # First 25 values for x
    y_coords = keypoints[num_keypoints:2*num_keypoints]  # Next 25 for y

    if len(x_coords) != len(y_coords):
        print("Warning: x and y sizes do not match!")
        return

    plt.scatter(x_coords, y_coords) # call scatter from plt
    plt.title(label)
    plt.axis('equal')
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical

# Encode labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# One-hot encode
y_train_encoded = to_categorical(y_train_encoded)
y_test_encoded = to_categorical(y_test_encoded)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout

# Define the model
model = Sequential()

# GRU layer (you can adjust units based on your needs)
model.add(GRU(units=128, return_sequences=False, input_shape=(75, 1)))

# Optional: Dropout layer to prevent overfitting
model.add(Dropout(0.5))

# Output layer (number of gestures/classes)
model.add(Dense(units=y_train_encoded.shape[1], activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
X_train_reshaped = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test_reshaped = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Define the EarlyStopping callback
early_stopping = EarlyStopping(
    monitor='val_loss',  # Monitor the validation loss
    patience=5,          # Number of epochs with no improvement after which training will be stopped
    restore_best_weights=True  # Restore weights from the epoch with the best validation loss
)

# Train the model with the EarlyStopping callback
history = model.fit(
    X_train_reshaped,
    y_train_encoded,
    epochs=50,
    batch_size=32,
    validation_data=(X_test_reshaped, y_test_encoded),
    callbacks=[early_stopping]  # Include the EarlyStopping callback
)
```

```python
import cv2
import numpy as np

def predict_gesture_from_video(video_path, model, label_encoder):
    cap = cv2.VideoCapture(video_path)
    frames = []
    count = 0

    while count < 30:
        ret, frame = cap.read()
        if not ret:
            break

        # Resize to 75x75 pixels
        frame = cv2.resize(frame, (75, 75))

        # Convert to grayscale
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Normalize the frame values to [0, 1]
        frame = frame / 255.0

        # Flatten the frame to fit the expected input for the model
        frame = frame.flatten()  # Shape: (5625,)

        # Append the flattened frame to the list
        frames.append(frame)
        count += 1

    cap.release()

    # Check how many frames were captured
    print(f"Number of frames captured: {len(frames)}")

    # If fewer than 30 frames, pad with zeros
    while len(frames) < 30:
        frames.append(np.zeros((75 * 75,)))  # Append zero array of shape (5625,)

    # Convert frames to a NumPy array
    input_frames = np.array(frames)  # Shape should be (30, 5625)

    # Reshape to fit the model's expected input shape (1, 30, 75, 75)
    input_for_prediction = input_frames.reshape(1, 30, 75, 75)  # Shape: (1, 30, 75, 75)

    # Now, make the predictions
    predictions = model.predict(input_for_prediction)  # Predictions for the batch

    # Get the predicted classes from predictions
    predicted_classes = np.argmax(predictions, axis=1)  # Get class indices

    # Determine the most common class across predictions
    predicted_class = np.argmax(np.bincount(predicted_classes))  # Most frequent prediction

    # Inverse transform to get the gesture label
    gesture = label_encoder.inverse_transform([predicted_class])

    return gesture[0]  # Return the first element of the array

# Usage example
video_path = '/content/test/test/best/s0002.mp4'  # Path to your video
gesture_result = predict_gesture_from_video(video_path, model, label_encoder)
print(f"Predicted gesture: {gesture_result}")
```

```python
def predict_gesture_real_time(model, label_encoder):
    cap = cv2.VideoCapture(0)  # Use 0 for webcam
    frames = []
    count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Preprocess the frame as above
        frame = cv2.resize(frame, (75, 75))
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        frame = frame.flatten() / 255.0  # Normalize

        frames.append(frame)
        count += 1

        if count == 30:  # Make predictions after collecting 30 frames
            input_frames = np.array(frames).reshape(1, 30, 75, 1)
            predictions = model.predict(input_frames)
            predicted_label = label_encoder.inverse_transform([predictions.argmax(axis=1)[0]])
            print("Predicted Gesture:", predicted_label[0])
            frames.clear()  # Clear frames to start collecting new ones
            count = 0  # Reset the counter

        # Display the frame (optional)
        cv2.imshow("Webcam Feed", frame.reshape(75, 75))  # Display resized grayscale frame

        if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to quit
            break

    cap.release()
    cv2.destroyAllWindows()

# Usage example for real-time prediction
predict_gesture_real_time(model, label_encoder)
```

```python
import cv2
import os

def extract_frames(video_path, save_path, num_frames=30):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    interval = total_frames // num_frames

    os.makedirs(save_path, exist_ok=True)

    frame_count = 0
    for i in range(num_frames):
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_count)
        ret, frame = cap.read()
        if ret:
            cv2.imwrite(f"{save_path}/frame_{i}.jpg", frame)
        frame_count += interval
    cap.release()

# Extract frames for train, test, and validate datasets
folders = ['/content/test/test/']
output_folders = ['/content/extracted_frames/test/']

for folder, output_folder in zip(folders, output_folders):
    for gesture_folder in os.listdir(folder):
        gesture_path = os.path.join(folder, gesture_folder)
        for video_file in os.listdir(gesture_path):
            video_path = os.path.join(gesture_path, video_file)
            extract_frames(video_path, f"{output_folder}/{gesture_folder}/{os.path.splitext(video_file)[0]}")
```

```python
import cv2
import os

def extract_frames(video_path, save_path, num_frames=30):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    interval = total_frames // num_frames

    os.makedirs(save_path, exist_ok=True)

    frame_count = 0
    for i in range(num_frames):
        cap.set(cv2.CAP_PROP_POS_FRAMES, frame_count)
        ret, frame = cap.read()
        if ret:
            cv2.imwrite(f"{save_path}/frame_{i}.jpg", frame)
        frame_count += interval
    cap.release()

# Extract frames for train, test, and validate datasets
folders = ['/content/validate/validate/']
output_folders = ['/content/extracted_frames/validate/']

for folder, output_folder in zip(folders, output_folders):
    for gesture_folder in os.listdir(folder):
        gesture_path = os.path.join(folder, gesture_folder)
        for video_file in os.listdir(gesture_path):
            video_path = os.path.join(gesture_path, video_file)
            extract_frames(video_path, f"{output_folder}/{gesture_folder}/{os.path.splitext(video_file)[0]}")
```

```python
import mediapipe as mp
import numpy as np

# Initialize MediaPipe Hands solution
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.5)

def extract_hand_keypoints(image_path):
    image = cv2.imread(image_path)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    results = hands.process(image_rgb)
    if results.multi_hand_landmarks:
        for hand_landmarks in results.multi_hand_landmarks:
            keypoints = np.array([[lm.x, lm.y, lm.z] for lm in hand_landmarks.landmark])
            return keypoints.flatten()
    return np.zeros(63)  # If no hand detected

def prepare_data(folder_path):
    keypoints_data = []
    labels = []

    for gesture_folder in os.listdir(folder_path):
        gesture_path = os.path.join(folder_path, gesture_folder)
        if not os.path.isdir(gesture_path):
            continue
        for frame in os.listdir(gesture_path):
            frame_path = os.path.join(gesture_path, frame)
            keypoints = extract_hand_keypoints(frame_path)
            keypoints_data.append(keypoints)
            labels.append(gesture_folder)  # Label = folder name (gesture)

    return np.array(keypoints_data), np.array(labels)

# Prepare data for training
train_folder = '/content/extracted_frames/train_data'
X_train, y_train = prepare_data(train_folder)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Encode labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

# Reshape data for GRU (batch_size, timesteps, features)
X_train = X_train.reshape(X_train.shape[0], 30, -1)  # 30 frames, 63 features

# Define the GRU model
model = Sequential([
    GRU(128, return_sequences=False, input_shape=(30, 63)),  # 30 timesteps, 63 keypoints
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(len(label_encoder.classes_), activation='softmax')  # Output layer for classification
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
history = model.fit(X_train, y_train_encoded, validation_split=0.2, epochs=50, batch_size=32)
```

```python
import mediapipe as mp
import cv2
import os
import numpy as np

# Initialize MediaPipe hands and pose detection
mp_hands = mp.solutions.hands
mp_pose = mp.solutions.pose
hands = mp_hands.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.5)
pose = mp_pose.Pose(static_image_mode=True)

def extract_keypoints(image_path):
    # Check if the image file is a valid image
    if not image_path.lower().endswith(('.png', '.jpg', '.jpeg')):
        print(f"Error: Not a valid image file: {image_path}")
        return np.zeros(75)  # Return a zero array for error cases

    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Could not read image from {image_path}")
        return np.zeros(75)  # Return a zero array for error cases

    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Extract hand landmarks
    hand_results = hands.process(image_rgb)
    hand_keypoints = np.zeros(63)  # 21 keypoints * 3 (x, y, z)
    if hand_results.multi_hand_landmarks:
        hand_keypoints = np.array([[lm.x, lm.y, lm.z] for lm in hand_results.multi_hand_landmarks[0].landmark]).flatten()

    # Extract pose landmarks (4 specific keypoints)
    pose_results = pose.process(image_rgb)
    pose_keypoints = np.zeros(12)  # 4 keypoints * 3 (x, y, z)
    if pose_results.pose_landmarks:
        pose_keypoints = np.array([[pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].z],
                                   [pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].z],
                                   [pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].z],
                                   [pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].z]]).flatten()

    return np.concatenate([hand_keypoints, pose_keypoints])


def load_test_data(folder_path):
    keypoints_data = []
    labels = []

    for gesture_folder in os.listdir(folder_path):
        gesture_path = os.path.join(folder_path, gesture_folder)
        if not os.path.isdir(gesture_path):
            continue
        for video_folder in os.listdir(gesture_path):  # Iterate through video folders
            video_path = os.path.join(gesture_path, video_folder)
            if not os.path.isdir(video_path):
                continue
            for frame in os.listdir(video_path):  # Now iterate through frames
                frame_path = os.path.join(video_path, frame)
                print(f"Processing frame: {frame_path}")  # Print each frame path
                keypoints = extract_keypoints(frame_path)
                keypoints_data.append(keypoints)
                labels.append(gesture_folder)  # Label = folder name (gesture)

    return np.array(keypoints_data), np.array(labels)

# Load test data from the appropriate directory
test_folder = '/content/drive/MyDrive/extracted_frames/test'  # Update this path to your test data
X_test, y_test = load_test_data(test_folder)

print(f"Test Features shape: {X_test.shape}, Labels shape: {y_test.shape}")

```

```python
# Step 1: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 2: Load the saved model
from tensorflow.keras.models import load_model

# Load the saved model from Google Drive
model_path = '/content/drive/MyDrive/saved_sign_language_model.h5'  # Update this path
model = load_model(model_path)

print("Model loaded successfully.")

# Assuming you have the training data
# If you have a function like `prepare_data`, you can re-run that for the training set:
# Example:
train_folder = '/content/drive/MyDrive/extracted_frames/train'
X_train, y_train = prepare_data(train_folder)

# Then fit the label encoder using y_train
label_encoder = LabelEncoder()
label_encoder.fit(y_train)  # Use training labels to fit the encoder

# Now, you can proceed with encoding y_test using the already fitted encoder
y_test_encoded = label_encoder.transform(y_test)

# Just to ensure y_train is defined
# You should already have y_train prepared during your training data preparation step
# For this example, we can assume some placeholder data for y_train.
# If you already processed it earlier, make sure to run that code block again.

# Example placeholder (this should be your actual training labels):
# y_train = np.array(['bad', 'best', 'glad', 'sad', 'scared', 'stiff', 'surprise'] * 100)  # Placeholder

# Step 4: Load Test Data
import mediapipe as mp
import cv2
import numpy as np
import os

# Initialize MediaPipe hands and pose detection
mp_hands = mp.solutions.hands
mp_pose = mp.solutions.pose
hands = mp_hands.Hands(static_image_mode=True, max_num_hands=2, min_detection_confidence=0.5)
pose = mp_pose.Pose(static_image_mode=True)

def extract_keypoints(image_path):
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error: Could not read image from {image_path}")
        return np.zeros(75)  # Return a zero array for error cases

    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Extract hand landmarks
    hand_results = hands.process(image_rgb)
    hand_keypoints = np.zeros(63)  # 21 keypoints * 3 (x, y, z)
    if hand_results.multi_hand_landmarks:
        hand_keypoints = np.array([[lm.x, lm.y, lm.z] for lm in hand_results.multi_hand_landmarks[0].landmark]).flatten()

    # Extract pose landmarks (4 specific keypoints)
    pose_results = pose.process(image_rgb)
    pose_keypoints = np.zeros(12)  # 4 keypoints * 3 (x, y, z)
    if pose_results.pose_landmarks:
        pose_keypoints = np.array([[pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].z],
                                   [pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].z],
                                   [pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_ELBOW].z],
                                   [pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].x,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].y,
                                    pose_results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_ELBOW].z]]).flatten()

    return np.concatenate([hand_keypoints, pose_keypoints])

def load_test_data(folder_path):
    keypoints_data = []
    labels = []

    for gesture_folder in os.listdir(folder_path):
        gesture_path = os.path.join(folder_path, gesture_folder)
        if not os.path.isdir(gesture_path):
            continue
        for video_folder in os.listdir(gesture_path):  # Iterate through video folders
            video_path = os.path.join(gesture_path, video_folder)
            if not os.path.isdir(video_path):
                continue
            for frame in os.listdir(video_path):  # Now iterate through frames
                frame_path = os.path.join(video_path, frame)
                keypoints = extract_keypoints(frame_path)
                keypoints_data.append(keypoints)
                labels.append(gesture_folder)  # Label = folder name (gesture)

    return np.array(keypoints_data), np.array(labels)

# Load test data from the appropriate directory
test_folder = '/content/drive/MyDrive/extracted_frames/test'  # Update this path to your test data
X_test, y_test = load_test_data(test_folder)

print(f"Test Features shape: {X_test.shape}, Labels shape: {y_test.shape}")

# Step 5: Encode Test Labels
from sklearn.preprocessing import LabelEncoder

# Fit the LabelEncoder on the test labels directly
label_encoder = LabelEncoder()
label_encoder.fit(y_test)  # Fit using the test labels

# Now encode y_test using the fitted encoder
y_test_encoded = label_encoder.transform(y_test)

# Proceed with evaluation


# Check unique encoded values to see if they match
print("Unique encoded labels in y_test:", np.unique(y_test_encoded))

# Prepare y_test_final to match the number of samples
num_frames = 30  # The number of frames used in your model
num_samples = len(X_test) // num_frames  # This gives the number of complete samples

# Create labels for each complete set of frames
y_test_final = y_test_encoded[:num_samples * num_frames]  # Only take the corresponding part
y_test_final = np.repeat(y_test_final, num_frames)[:len(X_test)]  # Repeat labels and truncate to match the number of features

# Check the sizes again
print(f"Reshaped X_test: {X_test.shape}, Reshaped y_test: {y_test_final.shape}")

# Ensure both shapes match for evaluation
assert len(X_test) == len(y_test_final), "Feature and label lengths do not match!"

# Reshape X_test for GRU if needed
X_test_reshaped = X_test.reshape(-1, num_frames, 75)

# Evaluate the model on the test data
test_loss, test_acc = model.evaluate(X_test_reshaped, y_test_final)
print(f"Test Accuracy: {test_acc}")

```
