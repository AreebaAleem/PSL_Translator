```python
# Load JSON data
with open(json_path) as f:
    data = json.load(f)
print(f"Loaded JSON data with {len(data)} entries.")

# Determine the number of unique classes dynamically
all_actions = [details['action'][0] for details in data.values()]
unique_actions = list(set(all_actions))
num_classes = len(unique_actions)
print(f"Determined num_classes: {num_classes}")

# Create a mapping for action labels to indices
action_to_index = {action: idx for idx, action in enumerate(unique_actions)}

# Function to find videos
def find_videos(data, subset="train", train_ratio=0.8):
    found_videos = []
    for video_id, details in data.items():
        if details['subset'] == subset:
            video_file = os.path.join(video_path, f"{video_id}.mp4")
            if os.path.exists(video_file):
                action_index = action_to_index[details['action'][0]]
                found_videos.append((video_file, action_index))
            else:
                print(f"Warning: Video file {video_file} does not exist.")
    print(f"Found {len(found_videos)} videos in subset '{subset}'")
    train_videos, val_videos = train_test_split(found_videos, train_size=train_ratio, random_state=42)
    print(f"Training set: {len(train_videos)} videos, Validation set: {len(val_videos)} videos")
    return train_videos, val_videos

# Function to extract frames from a video file
def extract_frames(video_file):
    cap = cv2.VideoCapture(video_file)
    frames = []
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        frame = cv2.resize(frame, frame_size)
        frames.append(frame)
    cap.release()
    return np.array(frames)

# Function to pad or trim frames
def pad_or_trim_frames(frames, max_frames):
    num_frames = len(frames)
    if num_frames < max_frames:
        padding = np.zeros((max_frames - num_frames, *frame_size, 3), dtype=np.uint8)
        frames = np.concatenate((frames, padding), axis=0)
    else:
        frames = frames[:max_frames]
    return frames

# Generator function to yield data in batches
def video_data_generator(videos, num_classes, batch_size=4):
    while True:
        X_batch = []
        y_batch = []
        for i in range(0, len(videos), batch_size):
            for video_file, action_index in videos[i:i + batch_size]:
                frames = extract_frames(video_file)
                if frames is None or len(frames) == 0:
                    continue
                frames = pad_or_trim_frames(frames, max_frames)
                frames = np.array(frames, dtype=np.float32) / 255.0
                label = to_categorical([action_index], num_classes=num_classes)
                X_batch.append(frames)
                y_batch.append(label)

            if X_batch:
                X_batch = np.array(X_batch)
                y_batch = np.array(y_batch)
                print(f"Debug (before yield): X_batch shape: {X_batch.shape}, y_batch shape: {y_batch.shape}")
                yield X_batch, y_batch

# Prepare data generators
train_videos, val_videos = find_videos(data, subset="train")
train_gen = video_data_generator(train_videos, num_classes=num_classes)
val_gen = video_data_generator(val_videos, num_classes=num_classes)
```

```python
import json
import os
import requests
import yaml

# Load glosses from data.yaml
yaml_path = '/content/drive/MyDrive/Dataset/data.yaml'
with open(yaml_path) as f:
    data_yaml = yaml.safe_load(f)

glosses_to_download = data_yaml['names']  # Assuming 'names' contains the glosses

# Load JSON data from WLASL
json_path = '/content/drive/MyDrive/Dataset/WLASL_v0.3.json'
with open(json_path) as f:
    data = json.load(f)

# Create directories for videos
train_dir = '/content/drive/MyDrive/Dataset/train/videos'
val_dir = '/content/drive/MyDrive/Dataset/val/videos'
test_dir = '/content/drive/MyDrive/Dataset/test/videos'
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Function to download videos
def download_video(url, save_path):
    try:
        r = requests.get(url, stream=True)
        r.raise_for_status()  # Check for request errors
        with open(save_path, 'wb') as f:
            for chunk in r.iter_content(chunk_size=8192):
                f.write(chunk)
        print(f'Downloaded: {save_path}')
    except Exception as e:
        print(f'Error downloading {url}: {e}')

# Iterate through the dataset to find and download relevant videos
for item in data:
    gloss = item['gloss']
    if gloss not in glosses_to_download:  # Filter glosses based on data.yaml
        continue

    instances = item['instances']
    for instance in instances:
        video_url = instance['url']
        split = instance['split']  # Determine if it's train, val, or test

        # Set the save directory based on the split
        if split == 'train':
            save_dir = train_dir
        elif split == 'val':
            save_dir = val_dir
        elif split == 'test':
            save_dir = test_dir
        else:
            print(f'Unknown split: {split} for gloss: {gloss}')
            continue

        # Download video to the respective directory without gloss folder
        video_filename = os.path.basename(video_url)
        video_save_path = os.path.join(save_dir, video_filename)
        download_video(video_url, video_save_path)

print("Download completed for all relevant videos.")
```

```python
import json
import os

# Paths to save the train, val, and test frame lists
train_txt_path = '/content/drive/MyDrive/Dataset/frames_train.txt'
val_txt_path = '/content/drive/MyDrive/Dataset/frames_val.txt'
test_txt_path = '/content/drive/MyDrive/Dataset/frames_test.txt'

# Load JSON data from WLASL to get split info
json_path = '/content/drive/MyDrive/Dataset/WLASL_v0.3.json'
with open(json_path) as f:
    data = json.load(f)

# Define frame directory path
frame_dir = '/content/drive/MyDrive/Dataset/frames'

# Open files to write paths
with open(train_txt_path, 'w') as train_file, \
     open(val_txt_path, 'w') as val_file, \
     open(test_txt_path, 'w') as test_file:

    # Iterate over each item in WLASL data to organize frames by split
    for item in data:
        gloss = item['gloss']
        instances = item['instances']

        for instance in instances:
            split = instance['split']
            video_id = instance['video_id']

            # Assuming frames are saved with video_id naming convention in frame_dir
            frame_path = os.path.join(frame_dir, f"{video_id}.jpg")  # Adjust extension if needed

            # Write frame path to the correct split file
            if split == 'train':
                train_file.write(frame_path + '\n')
            elif split == 'val':
                val_file.write(frame_path + '\n')
            elif split == 'test':
                test_file.write(frame_path + '\n')

print("Frame lists created successfully for train, val, and test splits.")
```
